## DataFrame

Un dataframe es una colección rectangular de variables (columnas) y observaciones (filas) `penguins` contiene 344 observaciones.

Definir algunos terminos
- Variable: Un cantidad, cualidad, o propiedad que se puede medir
- Valor: El estado de una variable cuando la mides. El valor de la variable puede cambiar de medición en medición
- Observación: Conjunto de mediciones realizadas en condiciones similares. Una observación contendrá varios valores, cada uno asociado a una variable diferente.
- Datos tabulados: Un conjunto de valores, cada uno asociado con una variable y una observación.

En este contexto, una variable refiere a un atributo de todos los pingüinos, y una observación refiere a todos los atributos de un solo pingüino.

## Creando un ggplot

Para crear un gráfico se comienza utilizando la función `ggplot()`, definiendo un objeto 'plot' al que se le agregarán capas (**layers**) al mismo. El primer argumento de `ggplot()` es el conjunto de datos que se usará para crear el gráfico `ggplot(data = penguins)` crea un gráfico vacio que se utilizará para mostrar los datos del conjunto `penguins`.

```python
ggplot(data = penguins)
```

El argumento `mapping` de la función `ggplot()` define como las variables del conjunto de datos se asignaran a las propiedades visuales (**aesthetics**) del gráfico.

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
  )
```

```python
# Definimos un objeto geometrico que nuestro gráfico usará para representar los datos
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
  ) +
    geom_point()
```

## Añadir estéticas y capas

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = species)
  ) +
    geom_point()
```

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = species)
) +
  geom_point() +
  geom_smooth(method = "lm")
```

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(mapping = aes(color = species)) +
  geom_smooth(method = "lm")
```

## `shape` aesthetic

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(mapping = aes(color = species, shape = species)) +
  geom_smooth(method = "lm")
```

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(aes(color = species, shape = species)) +
  geom_smooth(method = "lm") +
  labs(
    title = "Body mass and flipper length",
    subtitle = "Dimensions for Adelie, Chinstrap, and Gentoo Penguins",
    x = "Flipper length (mm)", y = "Body mass (g)",
    color = "Species", shape = "Species"
  )
```

```python
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(mapping = aes(color = bill_depth_mm)) +
  geom_smooth(method = "lm")
```

## Visualizando distribuciones

### Variables categóricas

Una variable es **categórica** si puede tomar uno o un pequeño grupo de valores.

```python
ggplot(penguins, aes(x = species)) +
  geom_bar()
```

```python
ggplot(penguins, aes(x = fct_infreq(species))) +
  geom_bar()
```

## Variables numéricas

Una variable es *numérica* (o cuantitativa) si puede tomar un amplio rango de valores numéricos y si se puede sumar, restar, o tomar promedios con esos valores. Las variables numéricas pueden ser continuas o discretas.

```python
ggplot(penguins, aes(x = body_mass_g)) +
  geom_histogram(binwidth = 200)
```

Esta función se utilizia para trazar una estimación de la densidad de probabilidad de un conjunto de datos. Esta función crea un gráfico de densidad suavizada que muestra la distribución de los datos en forma de una curva.

```python
ggplot(penguins, aes(x = body_mass_g)) +
  geom_density()
```

## Visualizing relationships

Una *boxplot* (diagrama de caja y bigotes) es un tipo de gráfico que mide la posición (cuartiles) que describen una distribución.

- La *caja* indica el rango de la mitad media de los datos, una distancia conocida como rango intecuartílico (IQR), que va desde el percentil 25 hata el 75. La linea dentro de la caja muestra la mediana, el percentil 50. Esto sirve para conocer la dispersión de la distribución y si ésta es simétrica respecto a la mediana, o sesgada hacia algún lado.
- Los puntos son observaciones que estan a más del 1.5 del rango intercuartilico (IQR). Estos puntos son conocidos como datos atípicos
- Las lineas (bigotes) se extienden de los dos lados de la caja y llegan hasta el punto más alejado de la distribución que no es un valor atípico.

```python
ggplot(penguins, aes(x = species, y = body_mass_g)) +
  geom_boxplot()
```

```python
ggplot(penguins, aes(x = body_mass_g, color = species)) +
  geom_density(linewidth = 0.75)
```

```python
ggplot(penguins, aes(x = body_mass_g, color = species, fill = species)) +
  geom_density(alpha = 0.5)
```

## Dos variables categóricas

Podemos utilizar una gráfica de barras apiladas para visualizar la relación entre dos variables categóricas.

```python
ggplot(penguins, aes(x = island, fill = species)) +
  geom_bar()
```

Esta es una gráfica de frecuencias relativas, es más util para comparar las distribuciones de las especias a traves de las islas

```python
ggplot(penguins, aes(x = island, fill = species)) +
  geom_bar(position = "fill")
```

## Tres o más variables numericas


```python
ggplot(penguins, aes(x = flipper_length_mm, y = body_mass_g)) +
  geom_point(aes(color = species, shape = island))
```

No es recomendable agregar muchas estéticas a un gráfico porque se hace dificil de leer. Otra función útil para las variables categóricas es dividir el gráfico en '*facets*'

```python
ggplot(penguins, aes(x = flipper_length_mm, y = body_mass_g)) +
  geom_point(aes(color = species, shape = species)) +
  facet_wrap(~island)
```
